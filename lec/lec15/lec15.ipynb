```python
def is_anagram(s1, s2):
    '''Returns True iff s1 and s2 are anagrams'''
    for c in s1:
        if c == " ":
            continue
        if c not in s2:
            return False
    
    for c in s2:
        if c == " ":
            continue
        if c not in s1:
            return False
    
    return True

def better(s1, s2):
    '''Returns True iff s1 and s2 are anagrams'''
    for c in s1 + s2:
        if c == " ":
            continue
        if s1.count(c) != s2.count(c):
            return False
    
    return True

def modified(s1, s2):
    s1 = list(s1)
    s2 = list(s2)
    x = set(s1)
    
    for i in x:
        if not s1.count(i) == s2.count(i):
            return False
    
    return True
```
